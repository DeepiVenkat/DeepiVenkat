## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

Check skewness and apply transformations

Convert data types if needed

Feature transformations (date parts, derived fields if required for analysis)

```python
# Find missing values
df.isnull().sum()
```

```python
# Fill numeric column with mean and median

df["Study_Hours_per_Week"].fillna(df["Study_Hours_per_Week"].median())
df["Department"].fillna(df["Department"].mode()[0])
```

```python
# Removing Duplicates
df.duplicated().sum()
```

```python
#correctin Datatypes
df["Stress_Level (1-10)"] = df["Stress_Level (1-10)"].astype(int)
df["Attendance (%)"] = df["Attendance (%)"].astype(float)
df
```

```python
#Creating Derived Columns  ==> creating Derived columns
df["Attendance_Category"] = pd.cut(
    df["Attendance (%)"],
    bins=[0, 60, 80, 100],
    labels=["Low", "Medium", "High"])

df
```

```python
#Pass/Fail from Grade.
df["Pass_Fail"] = df["Grade"].apply(lambda x: "Pass" if x != "F" else "Fail")
df
```

```python
print(df.columns)
#merging columns
df["Full_Name"] = df["First Name"] + " " + df["Last Name"]
cols = df.columns.tolist()
cols.insert(1, cols.pop(cols.index("Full_Name")))
df = df[cols]
df
```

```python
#average score by Departrment
dept_avg = df.groupby("Department")["Total_Score"].mean()
dept_avg
```

```python
#only student with >70% attendance
high_attendance = df[df["Attendance (%)"] > 70]
high_attendance
```

```python
#change gender in to “Male”,"Female" from "M","F"

df["Gender"] = df["Gender"].str.capitalize()
df["Gender"] = df["Gender"].replace({"M": "Male", "F": "Female"})
df
```

```python
df["Department"] = df["Department"].str.capitalize()
df["Department"] = df["Department"].replace({"CS": "Computer Science","Cs":"Computer Science"})
df
```

```python
#Consistency Checks
df = df[(df["Age"] >= 15) & (df["Age"] <= 30)]
df = df[(df["Attendance (%)"] >= 0) & (df["Attendance (%)"] <= 100)]
df = df[(df["Stress_Level (1-10)"] >= 1) & (df["Stress_Level (1-10)"] <= 10)]
```

```python
#reduce floating number in total score
df["Total_Score"] = df["Total_Score"].round(2)
df
```

```python
#round of sleep hours and study hours
df["Sleep_Hours_per_Night"] = df["Sleep_Hours_per_Night"].round(0).astype(int)
df["Study_Hours_per_Week"] = df["Study_Hours_per_Week"].round(0).astype(int)
df
```

```python
# roung of all score columns
df["Midterm_Score"]=df["Midterm_Score"].round(0).astype(int)
df["Final_Score"]=df["Final_Score"].round(0).astype(int)
df["Assignments_Avg"]=df["Assignments_Avg"].round(0).astype(int)
df["Total_Score"]=df["Total_Score"].round(0).astype(int)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# set style
sns.set(style="whitegrid")
```

```python
# Bivariate Analysis → relation between two variables
sns.barplot(x="Grade", y="Total_Score", data=df, estimator="mean", palette="viridis")
plt.title("Average Total Score by Grade")
plt.show()
```

What it shows:

       This chart shows the average performance of students across grades.

      Students in Grade A have the highest scores.

      Scores gradually decrease as grades drop (B → C → D → F).

      Clear evidence that academic grade aligns strongly with student performance.

```python
sns.barplot(x="Gender", y="Total_Score", data=df, estimator="mean", palette="pastel")
plt.title("Average Total Score by Gender")
plt.show()
```

```python
sns.barplot(x="Grade", y="Study_Hours_per_Week", hue="Gender", data=df, estimator="mean", palette="Set2")
plt.title("Average Study Hours by Grade and Gender")
plt.show()
```

What it Shows:

    This chart highlights the role of attendance in academic success.

    Students in higher grades maintain better attendance.

    Lower-grade students often have irregular attendance.

    Confirms that consistent attendance contributes to higher performance.

```python
sns.barplot(x="Grade", y="Attendance (%)", data=df, palette="coolwarm")
plt.title("Average Attendance % by Grade")
plt.show()
```

What it shows:

    Students in higher grades maintain better attendance.
